## Pandas data analysis

```python
import pandas as pd
```

```python
dataset_path = "../ai_jobs_market_2025_2026.csv"

df = pd.read_csv(dataset_path)

#check shape
print("Shape of dataset : rows X columns:", df.shape)
print("\nColumn names:\n", df.columns.tolist())
```

```python
# First and Last Rows
df.head(2)
#df.tail(2)
```

```python
print("\nInfo summary:", df.info())   # Column names, dtypes, and non-null counts
```

```python
# statistical ( Count, mean, std, min, max, percentiles) summary of all numeric columns
print("\nDescribe (numeric columns):")
print(df.describe())
```

```python

# Describe including all categorical/text columns
print("\nDescribe (all columns):")
print(df.describe(include="all"))
```

```python
#  Count missing values per column 
print("Missing values per column:")
print(df.isnull().sum())

# Total rows with atleast one missing value,  
rows_with_any_missing = df.isnull().any(axis=1).sum()
print(f"\nRows with at least one missing value: {rows_with_any_missing} ({rows_with_any_missing/len(df)*100:.1f}%)")
```

```python
# Check for duplicates
duplicate_count = df.duplicated().sum()
print(f"Fully duplicate rows: {duplicate_count}")
```

```python
# check if any row has salary_min_usd greater than annual_salary_usd
invalid_min = df[df["salary_min_usd"] > df["annual_salary_usd"]]
print("Rows where salary_min > annual_salary:", len(invalid_min))

invalid_max = df[df["annual_salary_usd"] > df["salary_max_usd"]]
print("Rows where annual_salary > salary_max:", len(invalid_max))

invalid_max = df[df["annual_salary_usd"] > df["salary_max_usd"]] # list few salaries > sallary_max
print(invalid_max[["job_id", "annual_salary_usd", "salary_min_usd", "salary_max_usd"]].head(10))
```
